# Problem Statement
- In this analysis, We will generate the Random Forest using the Cross Validation Splitting technique on IRIS Data 

```python
# Import inbuilt Iris Datset
from sklearn.datasets import load_iris
```

```python
# Load the dataset with dataframe
iris_data = load_iris()
print(iris_data)
```

```python
# Extract the onput data
data_input = iris_data.data
# Extract the target values
data_output = iris_data.target
print(data_input)
```

```python
print(data_output)
```

```python
# Using K-fold Cross validation techniques to split the dataset into 5 consecutive folds without shuffling bu default 

from sklearn.model_selection import KFold
kf = KFold(n_splits=5,shuffle=True)
```

```python
#Using a for loop we will noe iterating the individual train test sets with in the input data
```

```python
print("Train Set                  Test Set       ")
for train_set, test_set in kf.split(data_input):
    print(train_set,test_set)
```

```python
# Lets Initalize the Random forest Classifier 
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier(n_estimators=10)
```

```python
# Import the Cross Validation score liabrary
from sklearn.model_selection import cross_val_score 
print(cross_val_score(rf_class,data_input,data_output,scoring="accuracy",cv=10))
```

```python
accuracy = cross_val_score(rf_class,data_input,data_output,scoring="accuracy",cv=10).mean()*100
print("Accuracy of Random Forest is :", accuracy)
```
